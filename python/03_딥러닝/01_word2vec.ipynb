```python
# h = 활성함수(W이전h + Wx + B)
from gensim.models import Word2Vec
from konlpy.tag import Okt
import numpy as np
import pandas as pd
```

```python
# 문장 불러오기
# data/csv/sentiment_data_long_csv 불려와서 DataFrame로 변환
df = pd.read_csv('../data/csv/sentiment_data_long.csv')
# sentence만 추출
sentences = df['sentence'].values
```

```python
okt = Okt()
# 학교에 갔다
# 학교에 가가(19일차 MLP, RNN 예제에서 했던 방식)
# [(학교, 명사), (에, 조사), (가다, 동사)] 현재 예제
tonkende_sentences = [okt.morphs(s, stem=True) for s in sentences]
```

```python
model = Word2Vec(
	sentences=tonkende_sentences, # 학습 시킬 문장들
	vector_size=100, # 벡터화 시킬 차원 크기
	window=3, # 3단어씩
	min_count=1, # 최소 1번은 나와야 함
	sg = 1, # Skip-gram 적용
)
```

```python
def
before_word='음악'
after_word='울리다'

v_before = model.wv[before_word]
v_after = model.wv[after_word]
# 각 단어와 가까운 단어를 구하기 위해 평균을 냄
context_vector = (v_before + v_after) / 2.0
# 평균과 가장 유사한 단어를 가져옴
sim_words = model.wv.most_similar(positive=[context_vector], topn=5)

# 앞 단어와 뒷 단어는 제외
result = [word for word, _ in sim_words if word not in [before_word, after_word]]

# 결과를 출력
result[0]
```
