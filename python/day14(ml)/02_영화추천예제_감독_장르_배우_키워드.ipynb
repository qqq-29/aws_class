```python
import pandas as pd
import numpy as np

# csv파일들을 불러옴
df1 = pd.read_csv('../data/tmdb_5000_credits.csv')
df2 = pd.read_csv('../data/tmdb_5000_movies.csv')
# df1.head(5)
# df2.head(5)
print(df1.columns)
print(df2.columns)
df1.columns = ['id', 'title', 'cast', 'crew']
print(df1.columns)
# 두 csv파일을 하나로 합침(중복된 제목은 빼고)
df = df2.merge(df1[['id', 'cast', 'crew']], on='id')
df.columns
```

```python
from ast import literal_eval
# 장르의 값은 타입이 문자열
type(df['genres'][0])
# 장르의 값을 리스트로 변경
genre0 = literal_eval(df['genres'][0])
for genre in genre0:
	print(genre['name'])
```

```python
# cast의 값은 타입이 문자열
type(df['cast'][0])
# 장르의 값을 리스트로 변경
cast0 = literal_eval(df['cast'][0])
for cast in cast0:
	print(cast['name'])
```

```python
# keywords 값은 타입이 문자열
type(df['keywords'][0])
# keword의 값을 리스트로 변경
keyword0 = literal_eval(df['keywords'][0])
for keyword in keyword0:
	print(keyword['name'])
```

```python
# x데이터가 들어오면 x 데이터에서 name들만 추출하여 리스트를 만들어 반환
# 예 : cast 데이터가 들어오면 cast 데이터에서 배우 이름만 리스트로 만들어 반환
def get_list(x):
	# x가 리스트인자 확인
	if isinstance(x,list):
		# names = []
		# for i in x:
		# 	names.append(i['name'])
		names = [i['name']for i in x]
		# 처음부터 3개만 추출(최대 3개 추출)
		if len(names) > 3:
			names = names[:3] #0번지부터 3번지 전까지
		return names
	return []
```

```python
# genres, keywords, cast
get_list(literal_eval(df['keywords'][0]))
```

```python
features = ['genres', 'keywords', 'cast', 'crew']
# 문자열로된 값을 리스트로 변환 ""->[]
for feature in features:
	df[feature] = df[feature].apply(literal_eval)
```

```python
# 리스트 또는 문자열에 있는 공백을 제거하고 소문자로 만들어 주는 함수
def clean_data(x):
	if isinstance(x, list):
		return [str.lower(i.replace(' ','')) for i in x]
	elif isinstance(x, str):
		return str.lower(x.replace(' ',''))
	else:
		return ''
```

```python
for feature in features:
	df[feature] = df[feature].apply(clean_data)
```

```python
# crew 정보를 주면 감독을 추철
def get_director(x):
	for crew in x:
		if crew['job'] == 'Director':
			return crew['name']
	return np.nan
```

```python
def create_soup(x):
	return f'{' '.join(x['keywords'])} {' '.join(x['cast'])} {' '.join(x['genres'])} {x['director']}'
```

```python
df['soup'] = df.apply(create_soup, axis=1)
# df['soup']
```

```python
# 글자수를 계산
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer()
count_matrix = count.fit_transform(df['soup'])
```

```python
# 코사인 유사도 계산
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
indices = pd.Series(df.index, index=df2['title']).drop_duplicates()
def get_recommendations(title, cosine_sim=cosine_sim):
	# 영화 제목으로 idx를 찾음(왜? 해당 영화의 코사인 유사도를 가져오기 위해)
	# 코사인 유사도는 idx를 이용해서 찾아야 함
	idx = indices[title]
	# 코사인 유사도를 정렬하려고 하는데 그냥 정렬하면 기존 위치가 섞임
	# => 유사도만 보면 어떤 영화인지 모름
	# => 코사인 유사도를 (idx, 코사인유사도값)으로된 튜플을 만들면 섞여도 idx를 알수 있음
	sim_scores = list(enumerate(cosine_sim[idx])) # 영화 Avatar의 코사인 유사도
	sim_scores = sorted(sim_scores, key=lambda x : x[1], reverse=True)
	# 0번지는 검색영화이기 때문에 0번지를 제외한 10개 제출
	sim_scores = sim_scores[1:11]
	# 여기서 부터는 코사인 유사도 값이 필요없기 때문에 영화 idx만 추출해서 리스트로 만듬
	movie_indices = [i[0] for i in sim_scores]
	# 영화 idx를 이용해서 영화 제목들을 가져와서 반환
	return df['title'].iloc[movie_indices]
```

```python
print(get_recommendations('Avatar', cosine_sim))
```
