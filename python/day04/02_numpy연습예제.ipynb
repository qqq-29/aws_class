```python
# 홍길동 학생의 국어,영어,수학 성적을 콘솔 입력받아 numpy 배열로 만들고 콘솔에 출력하세요
# 콘솔에서 입력받을 때 사용할 문구 => 성적을 입력하세요 :

import numpy as np
subjects = ["국어", "영어", "수학"]
list = []
# 과목수만큼 반복문
for subject in subjects:
	# 성적을 입력
	num = input(f"{subject} : ")
	list.append(int(num))
scores = np.array(list)
print(scores)
```

```python
print(f"평군:{np.mean(scores)}")
print(f"평군:{scores.mean()}")
```

```python
# 판다스 예제
import pandas as pd
```

```python
# 리스트를 이용 => Series
list = [100, 20, 30, 40]
scores = pd.Series(list, index=["국어", "영어", "수학", "과학"])

print(scores)
# 딕셔너리를 이용 => DataFrame
scores_dic = {
	'과목' : ["국어", "영어", "수학", "과학"],
	'성적' : [100, 90, 80, 70],
	'출석' : [100,100,0,80],
	'수행' : [90,100,90,30]
}
df = pd.DataFrame(scores_dic)
print(df)
```

```python
# 데이터 탐색 예제
print(f"상위 2행 : \n{df.head(2)}")
print("-"*20)
print(f"하위 2행 : \n{df.tail(2)}")
print("-"*20)
print(f"info : \n{df.info()}")
print("-"*20)
print(f"describe : \n{df.describe()}")
print("-"*20)
print(f"shape : \n{df.shape}")
print("-"*20)
```

```python
def bar():
	print("-"*30)
# 데이터 선택 및 필터링
print("데이터 프레임")
bar()
print(df)
bar()

#과목들
print(df['과목'])
bar()

#국어 성적 선택
print(f"국어 성적 : \n{df.loc[0]}")
bar()
# 과목의 수행평가 점수 조회
print(f"과목과 수행 평가 조회")
# 조회하려는 열이 여러개인 경우 열 목록을 리스트로 전달
print(df[['과목', '수행']])
bar()

# 국어, 수학 정보 조회
print("국어, 수학 정보 조회")
print(df.loc[[0,2]])
bar()

# 국어, 수학 정보 조회
print("국어, 영어, 수학 정보 조회")
print(df.iloc[0:3])
bar()
# 국어정보 조회
print(f"국어 성정 : {df.loc[0, '성적']}")
bar()
# 성적이 80이상인 데이터를 조회
print(df[df['성적']>=80])
print(df.loc[df['성적']>=80])
bar()
# 성적이 80이상은 과목들을 조회
print(df[df['성적']>=80]['과목'])
print(df.loc[df['성적']>=80]['과목'])
#print(df.query('성적'>=80)['과목'])
```

```python
# 데이터 결합 및 그룹 예제
df3 = pd.DataFrame({
	'이름' : ['홍길동', '고길동', '둘리'],
	'국어성적' : [100, 90, 80],
})
df4 = pd.DataFrame({
	'이름' : ['홍길동', '고길동', '둘리'],
	'영어성적' : [80, 70, 90],
})
```

```python
# 국어 성적과 영어 성적 합치기
total = pd.merge(df3, df4, on="이름")
print(total)
```
