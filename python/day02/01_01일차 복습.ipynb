```python
# 정수를 입력받아 입력받은 정수가 홀수인지 짝수인지 출력하는 코드 => if문
a = input("정수를 입력하세요")
if int(a)%2==0:
    print("짝수")
if int(a)%2!=0:
    print("홀수")
a=int(a)

if a % 2==0:
    print(f"{a}는 짝수")
else:
    print(f"{a}는 홀수")

print(f"{a}는 {"짝수" if a % 2 == 0 else "홀수"}")
```

```python
# min부터 max까지의 수를 출력하는 함수를 만들고, 1부터 10까지 출력하세요
# 함수명 : printNums
def printNums(min, max):
    if min > max:
        min,max = max,min
    for i in range(min, max):
        print(i)
            
printNums(1, 10)
```
