```python
# 문장을 입력받아 각 단어가 몉번 사용됐는지 판별하는 코드를 작성하세요
# 입력 : 안녕 내 이름 홍길동 네 이름 뭐니 내 이름 임꺽정
text = "안녕 내 이름 홍길동 네 이름 뭐니 내 이름 임꺽정"

def analyze_words(sentence):
    list_text = sentence.split()
    dic = {}
    print(list_text)
    for word in list_text:
        if word not in dic:
            dic[word] = list_text.count(word)
    return(dic)
    
def print_dic(dic):
    for key, value in dic.items():
        print(f"\"{key}\" : {value}")

print_dic(analyze_words(text))
print("-------------------------------------------------")
print(f"안녕 : {list_text.count("안녕")}")
print(f"내 : {list_text.count("내")}")
print(f"이름 : {list_text.count("이름")}")
print(f"홍길동 : {list_text.count("홍길동")}")
print(f"네 : {list_text.count("네")}")
print(f"뭐니 : {list_text.count("뭐니")}")
print(f"임꺽정 : {list_text.count("임꺽정")}")
```
