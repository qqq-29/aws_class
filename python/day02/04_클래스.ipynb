```python
# 2차원 좌표평면의 점을 나타나는 Point 클랴서를 선언

# Point 클래스의 객체를 생성
class Point:
    name : str = "포인트 클래스"

    #생성자
    def __init__(self, x:int = 0, y:int = 0):
        #인스턴스 변수
        self.__x = x
        self.__y = y

    # 함수
    def print(self):
        print(f"({self.__x}, {self.__y})")

    @property # x의 getter
    def x(self):
        return self.__x

    @x.setter # x의 setter
    def x(self,x):
        self.__x = x

    @property # y의 getter
    def y(self):
        return self.__y

    @x.setter # y의 setter
    def y(self,y):
        self.__y = y
        
print(Point.name)
p1 = Point(0,0)
p2 = Point(1,1)
p3 = Point()
p1.print()
p2.print()
p3.x = 100
p3.y = 100
p3.print()
# 자바에서 setter사용
```

```python
# 은행 계좌 클래스
# 소유자(owner), 잔액(balance)
# 기능 : 출금(withdraw), 입금(deposit)

# 홍길동이 계좌 개설
class BankAccount:
    #생성자
    def __init__(self, owner : str, balance : int = 0):
        #인스턴스 변수
        self.__owner = owner
        self.__balance = balance

    @property
    def balance(self):
        print(f"{self.__owner}잔액: {self.__balance:,}원 입니다")

    def deposit(self, amount):
        self.__balance = self.__balance + amount
        print(f"{amount:,}원 입금")
        print(f"잔액: {self.__balance:,}원 입니다")

    def withdraw(self, amount):    
        if self.__balance < amount:
            print("잔액이 부족합니다")
        else:
            self.__balance = self.__balance - amount
            print(f"잔액: {self.__balance:,}원 입니다")

account = BankAccount("홍길동")
# 100000(십만) 입금
account.deposit(100000)
# 50000(5만) 출금
account.withdraw(50000)
# 70000(7만) 출금 => 잔액 부족 알림
account.withdraw(70000)
account.balance
```

```python
# 도서 대여
# 도서 클래스
# 저자(author), 도서명(title), 대출 기능 여부(is_borrow)
# 기능 : 대출, 반납, 정보확인
class Book:
    # 생성자
    def __init__(self, title, author, is_borrow):
        self.__title = title
        self.__author = author
        self.__is_borrow = is_borrow
        self.__borrower = ""

    # 대출
    def borrow_book(self, user_name : str):
        if self.__is_borrow is False:
            self.__is_borrow = True
            self.__borrower = user_name
            print(f"{user_name}님이 {self.__title}을 대출했습니다")
        else:
            print(f"{self.__title} 대출불가능")
        
    # 반납
    def return_book(self):
        self.__is_borrow = False
        print(f"{self.__borrower}님이 {self.__title} 반납성공")
        self.__borrower = ""
    # 정보확인
    def show_info(self):
        print(f"제목 : {self.__title}")
        print(f"저자 : {self.__author}")
        print(f"대출 여부 : {"대출 중" if self.__is_borrow else "대출 가능"}")
book1 = Book("파이썬 정석", "홀길동", False)
book2 = Book("AI 원리", "고길동", False)

book1.borrow_book("회원1") # 회원1님이 파이썬 정석을 대출했습니다.
book1.borrow_book("회원2") # 이미 대출 중인 도서라 대출이 불가능합니다.
book1.return_book() # 회원1님이 파이썬 정석을 반납했습니다. 해당도서는 대출이 가능합니다.

book1.show_info()
```
