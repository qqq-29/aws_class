```python
import os
# mnist 이미지를 저장할 폴더를 생성
os.makedirs('mnist_images', exist_ok=True)
```

```python
# mnist에서 제공하는 숫자 이미지를 다운로드
from sklearn.datasets import fetch_openml, load_breast_cancer, load_iris

# 데이터 로드. 숫자 이미지를 훈련 6만, 테스트 1만을 제공
mnist = fetch_openml('mnist_784', version=1, as_frame=False)

# 독립, 종속
X, y = mnist['data'], mnist['target']



```

```python
X[0]
```

```python
# 이미지 다운로드 # 꼭 필요한 코드는 아님. 생략 가능함
# 우리가 학습하는 이미지가 어떤 형태의 이미지인지 확인하기 위해 받음
import numpy as np
import cv2
# for X_tmp, y_tmp in zip(X, y):
for i in range(10000):
	# X는 1차원으로 되어 있어서 바로 이미지로 바꾸면 점이 찍힌 선이 됨
	# 그래서 원본인 28X28로 돌리기 위해 reshape을 함
	# image = X_tmp.reshape(28,28).astype(np.uint8)
	image = X[i].reshape(28,28).astype(np.uint8)
	label = y[i]
	# 0폴더~ 9폴더까지 생김(단, 저장하는 데이터가 너무 적으면 몇몇 폴더는 없을 수도 있음)
	# exist_ok=True 폴더있을떄 넘어가개 하는거
	os.makedirs(f'mnist_images/{label}', exist_ok=True)
	# 이미지 생성
	cv2.imwrite(f'mnist_images/{label}/img_{i}.png', image)
print("이미지 다운로드 완료")
```

```python
X_full, y_full = X, y
# 이미지를 학습시키기 위한 데이터로 변환
# (X) images : 1차행렬 리스트
# (y) labels : 1차행렬 리스트의 답
# images, labels = [], []
import glob
X, y = [], []

# 폴더별로 이미지들을 가져와서 X,y에 넣어주는 작업
# 폴더가 0폴더~9폴더이기 때문에 0부터 9까지 반복
for folder in range(10):
	# 폴더에 있는 이미지(.png)들을 가져와서 X,y에 넣어주는 작업
	file_path = f'mnist_images/{folder}/*.png'
	# 폴더에 있는 이미지들을 하나씩 꺼내서 작업
	for file in glob.glob(file_path):
		img = cv2.imread(file, cv2.IMREAD_GRAYSCALE)

		X.append(img.flatten())
		y.append(str(folder))

# 리스트를 numpy 리스트로 변환
X, y = np.array(X), np.array(y)
```

```python
# 훈련, 테스트 분리
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.svm import SVC
# 800개 데이터 학습
model = SVC(kernel='rbf', C=1.0)
model.fit(X_train, y_train)
```

```python
# 200개 데이터로 정확도 확인
model.score(X_test, y_test)
```

```python
# 7만개 데이터를 훈련,테스트로 분리
X_full_train, X_full_test, y_full_train, y_full_test = train_test_split(X_full,y_full, test_size=0.2)
#학습
model2 = SVC(kernel='rbf', C=1.0)
model2.fit(X_full_train, y_full_train)
```

```python
model2.score(X_full_test, y_full_test)
```

```python
model2.score(X_test, y_test)
```
