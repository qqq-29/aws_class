```python
# '취업'에 대해 학습 시키켜록한다
# 네이버 지식인에서 취업관 관련된 질문들을 가져와서 학습시키기 전 단계까지 처리하려고 한다

# 네이버 지식인에서 취업관 관련된 질문 200개를 추출해서 갸져오는 작업

# URL 지정

# 200개 가져올 때까지 반복 => 기본 10*20페이지

	# requests를 이용해서 응답을 요청

	# 질문을 추출 => 질문이 어디에 있는지
```

```python
# 추출한 데이터 전처히 작업
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

def job_naver(keyword : str, max_pages=20):
	titles = []
	hrefs = []

	headers = {
			'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Safari/537.36',
		}
	for page in range(1, max_pages+1):
		url = f"https://kin.naver.com/search/list.naver?query={keyword}&page={page}"

		response = requests.get(url, headers=headers)

		try:
			# 연결
			response.raise_for_status()
			# Beautifulsoup 객체 생성
			soup = BeautifulSoup(response.text, 'lxml')
			
			href_link = soup.select("a._searchListTitleAnchor")

			for q in href_link:
				title = q.text.strip()
				href = q.get("href")
				hrefs.append(href)
				titles.append(title)
					
		except Exception as e:
			print(f"예외 발생 : {e}")
	df = pd.DataFrame({
        "title": titles,
        "href": hrefs
    })
	# 중복 데이터 제거
	df = df.drop_duplicates(subset="href")
	print(df['href'][0])



# 결측치 처리

# 텍스트 길이 본석후 처리 : AI학습에서 짧은 문장은 정보량이 부족 => 제거
```
