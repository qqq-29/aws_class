```python
# 문자열.split(구본자)
# : 구분자를 기준으로 문자열 추출하여 리스트로 반환
str2 = "바나나 배 포도 사과"
fruits = str2.split()
print(fruits)

str3 = "바나나,뱌,포도,사과"
fruits = str3.split(",")
print(fruits)
```

```python
# 문자열.join(리스트)
# 리스트 앞에 문자열들을 추가하여 하나의 문자열로 만듬
tags = ["AI", "Python", "Web"]
result = " #".join(tags)
print(result) # AI #Python #Web
print(f"#{result}") # #AI #Python #Web
```

```python
# 문자열.startWith(A) : A로 시작하는지를 알려줌
# 문자열.endsWith(A) : A로 끝나는지를 알려줌
file_name = "report_2024.pdf"
print(f"{file_name}은 보고서입니까? {file_name.startswith("report")}")
print(f"{file_name}은 보고서입니까? {file_name.endswith(".pdf")}")
if file_name.endswith(".pdf"):
    print("PDF 파일입니다.")

if file_name.startswith("report"):
    print("보고서 형식의 파일입니다.")
```

```python
# 딕셔너리 선언
dic = {
    "name" : "홍길동",
    "age" : 21,
    "addr" : "서울시"
}
print(dic)
# 이름확인
print(f"이름 : {dic["name"]}") # 키가 없는 경우 예러 발생
#print(f"이름 : {dic["name1"]}") # 키가 없는 경우 예러 발생
print(f"이름 : {dic.get("name", "익명")}")
print(f"이름 : {dic.get("name1", "익명")}")# 키가 없어도 기본값이 실행
# 새로운 키 추가
dic["job"] = "강사"
print(dic)
```

```python
# set : {}를 선언
set_a = { "바나나", "사과", "딸기", "포도", "바나나" }
print(set_a)

set_b = { "바나나", "귤", "오렌지", "딸기" }
print(set_b)
```

```python
print("A와 B의 합잡합")
print(set_a | set_b) # "바나나", "사과", "딸기", "포도", "귤", "오렌지",
print(set_a.union(set_b))
```

```python
# 교집합 : &, intersection()
print("A와 B의 교잡합 : 두 집합에 공통으로 있는 값들")
print(set_a & set_b)
print(set_a.intersection(set_b))
```

```python
# 차지합 : -, difference()
print("A - B : 집합 B에 없고 A에만 있는 값들")
print(set_a - set_b)
print(set_a.difference(set_b))
```

```python
# 대칭 차집합 : ^ 
# 겹치지 않은 모든 원소들 모음
# (A - B) | (B - A)
# (A | B) - (A & B)
print(set_a ^ set_b) # (A - B) | (B - A)
```

```python
# 집합 예제 (조건, 이름은 중복X)
# 수학을 듣는 학생 목록
course_math = {"홍길동", "임꺽정", "둘리", "고길동", "하니"}
# 국어를 듣는 학생 목록
course_kor = {"이순신", "이도", "김삿갓", "둘리", "홍길동"}

# 수학만 듣는 학생 목록을 조회
print(f"수학만 듣는 학생 목록:{course_math - course_kor}")

# 수학, 국어 모두 듣는 학생 목록을 조회
print(course_math & course_kor)

# 수학 또는 국어를 듣는 학생 목록을 조회
print(course_math | course_kor)

# 수학이나 국어를 하나만 듣는 학생 목록을 조회
print(course_math ^ course_kor)
```
