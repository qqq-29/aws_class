```python
# num가 음수인지 양수인지 0인지를 판별하는 코드
num = 10
if num > 0:
    print(f"{num}는 양수")
elif num < 0:
    print(f"{num}는 음수")
else : 
    print(f"{num}는 0")
```

```python
# num가 홀수안지 짝수인지 판별
num = 11
match num % 2:
    case 0:
        print(f"{num}는 짝수")
    case 1:
        print(f"{num}는 홀수")
```
