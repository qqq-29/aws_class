```python
x = tf.Variable(3.0) # x = 3
# 기울기 구하기 y=x^2 함수에서 x가 3일 때 기울기를 구함
# y=x^2의 기울기함수는 y=2x(x^2을 미분하면)
# 함수 값 (y): x가 3일 때 x^2은 얼마인가? → 9.0
# 기울기 (dy/dx): x가 3일 때 "x를 아주 조금 변화시키면 $y$는 얼마나 빨리 변하는가?" → 6.0
with tf.GradientTape() as tape:
	y = x ** 2

dy_dx = tape.gradient(y, x)
print(dy_dx.numpy())
```

```python
# 키 데이터와 신발 데이터를 이용하여 학습하고, 키 데이터를 이용하여 신발 사이즈를 예측하는 예제
# 저수준으로 작성 => 직접 하나하나 작성
import tensorflow as tf
키데이터 = tf.constant([170, 180, 175, 160], dtype=tf.float32)
신발데이터 = tf.constant([260, 270, 265, 255], dtype=tf.float32)
```

```python
# 각 원소를 제곱
리스트 = [1.0,2.0,3.0]
제곱 = tf.square(리스트)
print(제곱)
제곱평균 = tf.reduce_mean(제곱)
print(제곱평균)
```

```python
# 신발 = w(가중치) * 키데이터 + b(편향)
w = tf.Variable(0.1) # 초기 임의의 값
b = tf.Variable(0.1) # 초기 임의의 값
# 옵티마이저 중 Adam 선택
opt = tf.keras.optimizers.Adam(learning_rate=0.1)

for i in range(1000):
	with tf.GradientTape() as tape:
		# 신발 = w(가중치) * 키데이터 + b(편향)
		예측값 = w * 키데이터 + b

		# 오차 : 평균 오차 제곱
		# 손실함수 중 평균 오차 제곱(MSE)를 선택해서 오차를 계산
		오차 = tf.reduce_mean(tf.square(신발데이터 - 예측값))
	# 가울기를 계산
	기울기 = tape.gradient(오차, [w,b])
	# 기울기를 이용하여 가중치 업데이터
	opt.apply_gradients(zip(기울기,[w,b]))

	if i % 100 == 0:
		print(f'{i}단계 : w = {w.numpy():.4f}, b = {b.numpy():.4f}, 오차 = {오차.numpy():.4f}')
```

```python
측정키 = 173
예측신발 = w.numpy() * 측정키 + b.numpy()
print(f'측정키 : {측정키}, 에측신발사이즈 : {예측신발:.2f}')
print(f'가중치 : {w.numpy():.2f}, 편향 : {b.numpy():.2f}')
```
