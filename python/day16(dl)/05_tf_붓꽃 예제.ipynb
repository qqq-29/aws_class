```python
# 붓꽃 예제를 딥러닝으로 예측
from tensorflow.keras import models, layers
import tensorflow as tf
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
# 붓꽃 데이터 생성
dataset = load_iris()
```

```python
# 독립, 종속 분리
X = dataset.data # 독립변수
y = dataset.target # 종속변수
```

```python
# 훈련, 테스트 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
```

```python
# 정규화
# 정규화를 하면 엄청 큰숫자,엄청 작은숫자들을 똑같이 '0~1 사이'나 '평균 근처'로 맞춰서, 
# 딱 큰숫자만 안보고
# AI가 모든 정보를 공평하게 보게 만듭니다.
nomalizer = layers.Normalization(axis=-1)
nomalizer.adapt(X_train)
```

```python
# 모델 설계
model = models.Sequential([
	nomalizer,
	layers.Dense(16, activation='relu'),
	layers.Dense(8, activation='relu'),
	layers.Dense(3, activation='softmax'),
])
```

```python
# 모델 설정
model.compile(optimizer='adam',
							loss='sparse_categorical_crossentropy',
							metrics=['accuracy'])
```

```python
# 학습
model.fit(X_train, y_train, epochs=1000, batch_size=16, verbose=0)
```

```python
# 실제(100번지) 6.3, 3.3, 6, 2.5 => 2
# 예측 6.4, 3.2, 6, 2.4 => 2
# 예측할 새로운 붓꽃
new_iris = np.array([[6.4, 3.2, 6, 2.4]], dtype='float32')
prediction = model.predict(new_iris)
```

```python
# 결과 출력
predic_res = np.argmax(prediction)
print(f'예측 품종[{predic_res}] : {dataset.target_names[predic_res]}')
```
