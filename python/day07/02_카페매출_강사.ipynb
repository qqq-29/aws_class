```python
# https://my-json-server.typicode.com/st8324/repo/cafe
# 카페 매출 정보를 가져와서 DataFrame으로 변환하는 코드를 작성하세요
import requests
import pandas as pd
def cafe()->pd.DataFrame:
	url = 'https://my-json-server.typicode.com/st8324/repo/cafe'

	response = requests.get(url)

	try:
		response.raise_for_status()
		# coffees = response.json()
		# list = []
		# labels = {
		# 	'주문시간' : '주문시간',
		# 	'메뉴' : '메뉴',
		# 	'카테고리' : '카테고리',
		# 	'수량' : '수량',
		# 	'가격' : '가격',
		# 	'결제수단' : '결제수단',
		# }
		# for coffee in coffees:
		# 	dic = {}
		# 	for column, label in labels.items():
		# 		dic[column] = coffee[label]
		# 	list.append(dic)
		return pd.DataFrame(response.json())
	except Exception as e:
		return pd.DataFrame()
```

```python
# 카페 매출 정보를 가져오는 함수
cafe()
```

```python
# 중복 및 결측치 처리하는 함수
def clean_cafe(df:pd.DataFrame)->pd.DataFrame:
	df = df.drop_duplicates()
	df = df.dropna()
	return df
```

```python
# 확인
df = cafe()
clean_cafe(df)
```

```python
# 데이터를 활용한 예제
# 매출을 조회하기 위해 '결제액'을 추가
# 결제액은 가격 * 수량
df['결제액'] = df['가격']*df['수량']
print(df)
```

```python
# 전체 매출액
revenue = df['결제액'].sum()
print(f'전체 매출액 : {revenue:,}원')
```

```python
# 메뉴별 매출액을 조회(메뉴, 수량, 매출액)
# groupby
menu_df = df.groupby("메뉴")[['수량','결제액']].sum()
print(menu_df)
```

```python
# 카테고리별 수량과 결제액을 조회
# category_df = df.groupby('카테고리')[['수량','결제액']].sum()
# category_df = category_df.sort_values(by='결제액', ascending=False)
category_df = df.groupby('카테고리')[['수량','결제액']].sum().sort_values(by='결제액', ascending=False)

# 카테고리별 매출 비율을 추가
# 전체 매출액 : revenue
print(category_df)
category_df['매출비율'] = 100*category_df['결제액']/revenue
category_df['매출비율'] = category_df['매출비율'].round(1) # 소수점 두번째 자리에서 반올림
print(category_df)

# 매출이 가장 높은 카테고리를 조회
max_revenue = category_df['결제액'].max()
print(category_df.loc[category_df['결제액']==max_revenue])

print("--------")
max_revenue_idx = category_df['결제액'].idxmax() # 결재액 중 가장 큰 값을 가지는 위치
print(f'매출이 가장 많은 카테고리 : {max_revenue_idx}')
print("--------")
print(category_df.loc[max_revenue_idx])
```

```python
# 시간대 별 주문 건수를 조회
# 시간대를 추가

df['주문시간'] = pd.to_datetime(df['주문시간'])
df['시간'] = df['주문시간'].dt.hour
print(df)

# 시간대 별 주몬 건수를 조회
hour_df = df['시간'].value_counts()

print(hour_df)
```

```python
# 시간대별로 매출액을 조회 : 오전(12시 이전), 오후(17시 이전), 저녁(17시 이후)
# 사간을 시간대로 변환하는 함수
def get_time_slot(hour):
	if hour < 12:
		return "오전"
	elif hour < 17:
		return "오후"
	else:
		return "저녁"

# 시간대를 출력
df['시간대'] = df['시간'].apply(get_time_slot)
print(df)

# 시간대 별 결제액을 조회
time_df=df.groupby('시간대')['결제액'].sum()
print(time_df)
```

```python
# 요일별 결제액 조회
df['요일'] = df['주문시간'].dt.dayofweek # dayofweek => 월 : 0, 화 : 1, .., 일 : 6
print(df)
```

```python
df['주문시간'] = pd.to_datetime(df['주문시간'])
df['요일'] = df['주문시간'].dt.day_name()
dayweek_map = {
	'Monday' : '월요일',
	'Tuesday' : '화요일',
	'Wednesday' : '수요일',
	'Thursday' : '목요일',
	'Friday' : '금요일',
	'Satday' : '토요일',
	'Sunday' : '일요일',
}
df['요일'] = df['요일'].map(dayweek_map)
print(df)
# 카테고리 리스트
weekday_list = ['월요일','화요일','수요일','목요일','금요일','토요일','일요일']

# 요일을 카테고리 타입으로 변환
df['요일'] = pd.Categorical(df['요일'], categories=weekday_list, ordered=True)

# 요일별 결제액을 조회
weekday_df = df.groupby('요일')['결제액'].sum()

# 정렬
weekday_df = weekday_df.sort_index()

print(weekday_df)
```

```python
# 전체를 카페_매출.csv로 저장
df.to_csv('카페_매출.csv', index=False)

# 메뉴별 매출을 카페_메뉴_매출.csv로 저장
menu_df.to_csv('카페_메뉴_매출.csv')

# 카테고리별 매출을 카페_카테고리_매출.csv로 저장
category_df.to_csv('카페_카테고리_매출.csv')

# 시간대별 매출을 카페_시간_매출.csv로 저장
hour_df.to_csv('카페_시간_매출.csv')

# 요일별 매출을 카페_요일_매출.csv로 저장
weekday_df.to_csv('카페_요일_매출.csv')
```
