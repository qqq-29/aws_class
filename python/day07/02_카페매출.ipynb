```python
# https://my-json-server.typicode.com/st8324/repo/cafe
# 카페 매출 정보를 가져와서 DataFrame으로 변환하는 코드를 작성하세요
import requests
import pandas as pd
def cafe()->pd.DataFrame:
	url = 'https://my-json-server.typicode.com/st8324/repo/cafe'

	response = requests.get(url)

	try:
		response.raise_for_status()
		coffees = response.json()
		list = []
		labels = {
			'주문시간' : '주문시간',
			'메뉴' : '메뉴',
			'카테고리' : '카테고리',
			'수량' : '수량',
			'가격' : '가격',
			'결제수단' : '결제수단',
		}
		for coffee in coffees:
			dic = {}
			for column, label in labels.items():
				dic[column] = coffee[label]
			list.append(dic)
		return pd.DataFrame(list)
	except Exception as e:
		return pd.DataFrame()
```

```python
# 카페 매출 정보를 가져오는 함수
cafe()
```

```python
# 중복 및 결측치 처리하는 함수
def clean_cafe(df:pd.DataFrame)->pd.DataFrame:
	df = df.drop_duplicates()
	df = df.dropna()
	return df
```

```python
# 확인
df = cafe()
clean_cafe(df)
```
