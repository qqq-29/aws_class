```python
import tensorflow as tf
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np

# 1. 사전 학습된 모델 로드 (ImageNet 데이터로 학습된 모델)
# ImageNet에는 약 1,000가지 사물(다양한 음식 포함) 정보가 들어있습니다.
model = MobileNetV2(weights='imagenet')

def predict_food(img_path):
    # 2. 이미지 불러오기 및 크기 조정 (224x224)
    img = image.load_img(img_path, target_size=(224, 224))
    
    # 3. 이미지를 배열로 변환하고 모델 형식에 맞게 전처리
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)

    # 4. 예측 수행
    preds = model.predict(x)
    
    # 5. 결과 해석 (상위 3개 결과 출력)
    print('예측 결과:')
    for (i, (id, label, prob)) in enumerate(decode_predictions(preds, top=3)[0]):
        print(f"{i+1}. {label}: {prob*100:.2f}%")

# 실행
# predict_food('pizza.jpg')
# predict_food('버거.jpg')
# predict_food('떡볶이.jpg')
# predict_food('오이.jpg')
# predict_food('구운오리.jpg')
# predict_food('치킨.jpg')
# predict_food('아이스크림.jpg')
# predict_food('빵.jpg')
# predict_food('고기국.jpg')
# predict_food('김치.jpg')
# predict_food('수시.jpg')
# predict_food('김밥.jpg')
predict_food('사과.jpg')
```
