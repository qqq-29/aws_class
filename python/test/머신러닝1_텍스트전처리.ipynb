```python
from konlpy.tag import Okt

okt = Okt()
stopwords = ['의', '가', '이', '은', '들', '는', '좀', '잘', '걍', '과', '도', '를', '으로', '자', '에', '와', '한', '하다']

raw_data = [
    "정말 최고의 영화였어요! 감동적입니다.",
    "영화가 너무 재미있어서 하나도 안 지루해요.",
    "기대 많이 했는데 생각보다 너무 별로네요.",
    "배우 연기는 별로지만 스토리가 살렸습니다."
]

def preprocess_korean(text):
    # 정규화
    tokens = okt.morphs(text, stem=True) 
    # 재조합
    clean_tokens = [word for word in tokens if word not in stopwords]
    return " ".join(clean_tokens)

preprocessed_data = [preprocess_korean(sentence) for sentence in raw_data]
print("결과:", preprocessed_data)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# TF-IDF 벡터라이저 생성
tfidf_vectorizer = TfidfVectorizer()

# 데이터 학습
tfidf_matrix = tfidf_vectorizer.fit_transform(preprocessed_data)
print(f"희소 행렬(Sparse Matrix)의 크기(Shape): {tfidf_matrix.shape}")
```
