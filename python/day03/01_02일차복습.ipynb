```python
# Phone 클래스를 추가
# 모델명(model), 가격(price)
# 기능 : 생성자 - 모델명과 가격을 초기화, 모델조회(display)

# IPhone 클래스를 추가(Phone 클래스를 상속)
# 페이스아이디적용여부(face_id)
# 기능 : 생성자 - 모델명과 가격, 페이스아이디 적용 여부를 초기화,
#				모델조회(display) - 모델명, 가격, 페이스적용여부

# Galaxy 클래스를 추가(Phone 클래스를 상속)
# 펜 지원(s_pen)
# 기능 : 생성자 - 모델명, 가격, 펜지원여부를 초기화
# 			모델조회(display) - 모델명, 가격, 페지원 여부

# IPhone : iPhone15, 1250000, True
# IPhone : iPhoneSE, 650000, False
# Galaxy : S24 Ultra, 1690000, True
# Galaxy : Z Flip 5, 1390000, False

# 각 핸드폰을 조회

class Phone:
	def __init__(self,model : str, price : int ):
		self.__model = model
		self.__price = price
	def display(self):
		print(f"모델명 : {self.__model}")
		print(f"가격 : {self.__price:,}원")

class IPhone(Phone):
	def __init__(self,model : str, price : int, face_id : bool):
		super().__init__(model, price)
		self.face_id = face_id

	def display(self):
		line_num = 20
		print("-" * line_num)
		super().display()
		print(f"페이스적용여부 : {"o" if self.face_id else "x"}")
		print("-" * line_num)

class Galaxy(Phone):
	def __init__(self,model : str, price : int, s_pen : bool):
		super().__init__(model, price)
		self.s_pen = s_pen

	def display(self):
		line_num = 20
		print("-" * line_num)
		super().display()
		print(f"펜 지원 : {self.s_pen}")
		print("-" * line_num)

I1 = IPhone("iPhone15", 1250000, True)
I1.display()
I2 = IPhone("iPhoneSE", 650000, False)
I2.display()
G1 = Galaxy("S24 Ultra", 1690000, True)
G1.display()
G2 = Galaxy("Z Flip 5", 1390000, False)
G2.display()
print("********************************************")
Phones = [
	IPhone("iPhone15", 1250000, True),
	IPhone("iPhoneSE", 650000, False),
	Galaxy("S24 Ultra", 1690000, True),
	Galaxy("Z Flip 5", 1390000, False),
]

for phone in Phones:
	phone.display()
```
