```python
# hearder에 User-Agent 추가
import requests

url = "https://jsonplaceholder.typicode.com/todos/1"

# 브라우저 정보 설정
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Safari/537.36'
}

response = requests.get(url, headers=headers)

if response.status_code == 200:
	print("연결 성공!")
	print(response.text)
```

```python
# URL에 파라미터 전달 예제
# ?name=홍길동&age=21

import requests

# URL에 접근하여 응답 객체 가져오기
url = "https://search.naver.com/search.naver"
param_dic = {'query': '파이썬'}
response = requests.get(url, params=param_dic)

# URL에 param을 확인
print(response.url)
print(response.text)
```

```python
# json 데이터 받아서 활용하기
import requests

url = "https://jsonplaceholder.typicode.com/todos/1"

response = requests.get(url)

if response.status_code == 200:
	print("연결 성공!")
	data = response.json()
	print(data)
	print("-"*30)
	for key, value in data.items():
		print(f"{key} : {value}")
```

```python
# 작은 파일 다운로드
import requests
# 고양이 이미지(구글에서 검색함)
url = "https://cdn.dailyvet.co.kr/wp-content/uploads/2024/05/15231624/20240515ceva_experts3.png"
response = requests.get(url)

# 응답이 성공적(200 OK)이라면 저장
if response.status_code == 200:
  with open("cat.jpg", "wb") as f: 
    f.write(response.content)
```

```python
# 큰 파일 다운로드
import requests
# 픽사베이에서 제공하는 무료 영상
url = "https://cdn.pixabay.com/video/2026/03/08/338904_large.mp4"

# stream=True 설정으로 연결만 먼저 열기
with requests.get(url, stream=True) as r:
  r.raise_for_status() # 에러 발생 시 예외 처리
    
  with open("video.mp4", "wb") as f:
    # 8KB씩 조각내서(chunk) 가져오기
      for chunk in r.iter_content(chunk_size=8192):
        f.write(chunk)

print("다운로드 완료!")
```

```python
import requests

url = "https://www.google.com"
response = requests.get(url)

if response.status_code == 200:
  # HTML 소스 코드 출력
  html_data = response.text
  print(html_data[:500])  # 너무 길어서 앞부분 500자만 출력
    
  # 파일로 저장
  with open("index.html", "w", encoding="utf-8") as f:
    f.write(html_data)
  print("저장완료")
```
